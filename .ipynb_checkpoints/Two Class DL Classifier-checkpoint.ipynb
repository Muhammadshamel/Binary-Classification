# Percptron Algorithm 
## Binary Classifier 

### Importing Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading data

```python
Data = pd.read_csv('./data.csv')
Data.head()
```

### Plotting

```python
cm = plt.cm.get_cmap('bwr')
label=Data['Y']
plt.scatter(data=Data , x='X1',y='X2',cmap=cm, c=label ) ;
```

### Prediction Function

```python
# Function to return the predicition in 1 or 0
def stepFunction(t):
    if t >= 0:
        return 1
    return 0

# Function calculates matrix multi with the points and weights + bias 
def prediction(X, W, b):
    return stepFunction((np.matmul(X,W.T)+b))
```

### Trigger Prediction

```python
# Assumed random Weights [W1 , W2]
W = np.array(np.random.rand(1,2))
# Assumed random bais
b = np.random.randint(-1,1)

# Loop to loop for all rows in Data 
for row in Data.values:
    
    # Call prediction function given (rows , Weights, bais)
    predict = prediction(row[:2] ,W ,b)
    
    print("Expected=%d, Predicted=%d" % (row[-1], predict))      
```

### Training Function 

```python
# Estimate Perceptron weights using stochastic gradient descent
def train_weights(train, l_rate, n_epoch):
   
    #Assume weights by zeros
    W = np.zeros(len(train[0])-1)
    #Assume random baises
    b = np.random.randint(-1,2)
    
    for epoch in range(n_epoch):
        sum_error = 0.0
        
        for row in train:
            
            #Call prediction function 
            predicted = prediction(row[:2], W, b)
            
            #Calculate the error
            error = row[-1] - predicted
            sum_error += error**2
            
            #Update bais
            b = b + l_rate * error
            
            for i in range(len(row)-1):
                #Update weights
                W[i] = W[i] + l_rate * error * row[i]
                
        print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error))
    return W,b
```

### Trigger Training 

```python
l_rate = 0.01
n_epoch = 1000
W,b = train_weights(Data.values, l_rate, n_epoch)

print('weights=',W,', bais=',b)
```

### Test Training

```python
for row in Data.values:
    # Call prediction function given (rows , Weights, bais)
    predict = prediction(row[:2], W, b)
    
    print("Expected=%d, Predicted=%d" % (row[-1], predict))     
```

# Boundry line with respect to the new weights and bais
We convert this equation w[0]X + w[1]y + b = 0  to this equation y= mx +c  
by moving the y term to the other side by negative sign then divied the other side with the coefficient of y 

```python
x = np.linspace(0, 1, 1000)

#Equation of line of seperation y= mx +c 
y = (x * W[0] + b) / (-1*(W[1]))

plt.scatter(data=Data , x='X1',y='X2',cmap=cm, c=label ) 
plt.plot(x, y, c= "yellow", marker='.', linestyle=':')


plt.show()
```
